## Collaborative Filtering

```python
book_ratings = ratings.merge(books,on='ISBN')

#experienced users are those who rate atleast 200 books
x = book_ratings.groupby('User-ID').count()['Book-Rating']>200
exp_users = x[x].index
filtered_users =  book_ratings[book_ratings['User-ID'].isin(exp_users)]
filtered_users
```

```python
#Books which are famous means they are atleast rated by 50 users
y = filtered_users.groupby('Book-Title').count()['Book-Rating']>50
famous_books = y[y].index

final_df = filtered_users[filtered_users['Book-Title'].isin(famous_books)]
final_df.head(10)
```

```python
#Mapping user id and isbn unique values to its index
user_unique = final_df['User-ID'].unique()
userid_to_idx = {user: id for id,user in enumerate(user_unique)}

isbn_unique = final_df['ISBN'].unique()
isbn_toidx = {isbn:id for id,isbn in enumerate(isbn_unique)}

final_df['user unique'] = final_df['User-ID'].map(userid_to_idx)
final_df['isbn unique'] = final_df['ISBN'].map(isbn_toidx)
```
